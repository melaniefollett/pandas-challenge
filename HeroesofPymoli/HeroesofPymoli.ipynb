```python
# Dependencies and Setup
import pandas as pd

# File to Load (Remember to Change These)
file_to_load = "purchase_data.csv"

# Read Purchasing File and store into Pandas data frame
purchase_df = pd.read_csv(file_to_load)
purchase_df.head()
```

```python
#Player Count

#Total number of players
total_players = len(purchase_df["SN"].unique())
total_players
```

```python
#Purchasing Analysis (Total)

#Number of Unique Items
num_items = len(purchase_df["Item ID"].unique())

#Average Purchase Price
avg_price = purchase_df["Price"].mean()

#Total Number of Purchases
total_purchases = len(purchase_df["Purchase ID"])

#Total Revenue
total_revenue = purchase_df["Price"].sum()
```

```python
#Create dataframe to store Purchasing Analysis (Total) data
summary_df = pd.DataFrame({
    "Number of Unique Items": [num_items],
    "Average Price": [avg_price],
    "Number of Purchases": [total_purchases],
    "Total Revenue": [total_revenue]
})

#Format dataframe
summary_df["Average Price"] = summary_df["Average Price"].astype(float).map("${:,.2f}".format)
summary_df["Total Revenue"] = summary_df["Total Revenue"].astype(float).map("${:,.2f}".format)

summary_df
```

```python
#Gender Demographics

#Remove multiple purchases by the same players
unique_SN_df = purchase_df.drop_duplicates(subset="SN")
unique_gender_df = unique_SN_df.groupby(unique_SN_df["Gender"])


#Percentage and Count of Male Players
male_total = unique_gender_df["Purchase ID"].count()["Male"]
male_percent = (male_total/total_players)*100

#Percentage and Count of Female Players
female_total = unique_gender_df["Purchase ID"].count()["Female"]
female_percent = (female_total/total_players)*100

#Percentage and Count of Other / Non-Disclosed
other_total = unique_gender_df["Purchase ID"].count()["Other / Non-Disclosed"]
other_percent = (other_total/total_players)*100

#Place data in dataframe 
gender_data = [{"Total Count": male_total, "Percentage of Players":male_percent},
                {"Total Count": female_total, "Percentage of Players":female_percent}, 
               {"Total Count": other_total, "Percentage of Players":other_percent}] 
gender_counts_df = pd.DataFrame(gender_data, index =["Male", "Female", "Other/Non-Disclosed"])

#Format dataframe
gender_counts_df["Percentage of Players"] = gender_counts_df["Percentage of Players"].astype(float).map("{:,.2f}%".format)

gender_counts_df
```

```python
#Purchasing Analysis (Gender)

#Use full dataframe with multiple purchases by same players, grouped by gender
gender_purchases_df = purchase_df.groupby(purchase_df["Gender"])

#Purchase Count by Gender
purchase_count = gender_purchases_df["Purchase ID"].count()

#Average Purchase Price by Gender
avg_purchase = gender_purchases_df["Price"].mean()

#Total Purchase Value by Gender
total_purchase = gender_purchases_df["Price"].sum()

#Average Purchase Total per Person by Gender
gender_count = unique_gender_df["Purchase ID"].count()
avg_purchase_individual = (total_purchase/gender_count)
```

```python
#Create dataframe to store Purchasing Analysis (Gender) data
purchase_gender_df = pd.DataFrame({
    "Purchase Count": purchase_count,
    "Average Purchase Price": avg_purchase,
    "Total Purchase Value": total_purchase,
    "Avg Total Purchase per Person": avg_purchase_individual
})

#Format dataframe
purchase_gender_df["Average Purchase Price"] = purchase_gender_df["Average Purchase Price"].astype(float).map("${:,.2f}".format)
purchase_gender_df["Total Purchase Value"] = purchase_gender_df["Total Purchase Value"].astype(float).map("${:,.2f}".format)
purchase_gender_df["Avg Total Purchase per Person"] = purchase_gender_df["Avg Total Purchase per Person"].astype(float).map("${:,.2f}".format)

purchase_gender_df
```

```python
#Age Demographics

#Establish bins for age
bins = [0, 9, 14, 19, 24, 29, 34, 39, 100]
group_names = ["<10", "10-14", "15-19", "20-24", "25-29", "30-34", "35-39", "40+"]

#Categorize the existing unique players using the age bins
unique_SN_df["Age Ranges"] = pd.cut(unique_SN_df["Age"], bins, labels=group_names)
unique_age_df = unique_SN_df.groupby(unique_SN_df["Age Ranges"])

#Calculate number and percentages by age group
unique_age_count = unique_age_df["Purchase ID"].count()
unique_age_percent = (unique_age_count/total_players)*100
```

```python
#Create dataframe to store Age Demographics data
count_age_df = pd.DataFrame({
    "Total Count": unique_age_count,
    "Percentage of Players": unique_age_percent,
})

#Format dataframe
count_age_df["Percentage of Players"] = count_age_df["Percentage of Players"].astype(float).map("{:,.2f}%".format)

count_age_df
```

```python
#Purchasing Analysis (Age)

#Establish bins for age
bins = [0, 9, 14, 19, 24, 29, 34, 39, 100]
group_names = ["<10", "10-14", "15-19", "20-24", "25-29", "30-34", "35-39", "40+"]

#Categorize the existing players (including multiple purchases) using the age bins
purchase_df["Age Ranges"] = pd.cut(purchase_df["Age"], bins, labels=group_names)
age_df = purchase_df.groupby(purchase_df["Age Ranges"])

#Purchase Count by Age
purchase_count_age = age_df["Purchase ID"].count()

#Average Purchase Price by Age
avg_purchase_age = age_df["Price"].mean()

#Total Purchase Value by Age
total_purchase_age = age_df["Price"].sum()

#Average Purchase Total per Person by Age
age_count = unique_age_df["Purchase ID"].count()
avg_purchase_person_age = (total_purchase_age/age_count)
```

```python
#Create dataframe to store Purchasing Analysis (Age) data
purchase_age_df = pd.DataFrame({
    "Purchase Count": purchase_count_age,
    "Average Purchase Price": avg_purchase_age,
    "Total Purchase Value": total_purchase_age,
    "Avg Total Purchase per Person": avg_purchase_person_age
})

#Format dataframe
purchase_age_df["Average Purchase Price"] = purchase_age_df["Average Purchase Price"].astype(float).map("${:,.2f}".format)
purchase_age_df["Total Purchase Value"] = purchase_age_df["Total Purchase Value"].astype(float).map("${:,.2f}".format)
purchase_age_df["Avg Total Purchase per Person"] = purchase_age_df["Avg Total Purchase per Person"].astype(float).map("${:,.2f}".format)

purchase_age_df
```
